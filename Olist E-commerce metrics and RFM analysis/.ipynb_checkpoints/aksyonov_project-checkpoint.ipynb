## 1) Downloading Data via Yandex-API

```python
def get_data(url):
    """ Function to download data from Yandex API """
    base_url = 'https://cloud-api.yandex.net/v1/disk/public/resources/download?'
    final_url = base_url + urlencode(dict(public_key=url))
    response = requests.get(final_url) 
    
    download_url = response.json()['href']
    df = pd.read_csv(download_url, sep=',')
    return df
```

```python
customers_url = 'https://disk.yandex.ru/d/ckMKdQdqChiPtQ'
orders_url = 'https://disk.yandex.ru/d/0zWmh1bYP6REbw'
order_items_url = 'https://disk.yandex.ru/d/1I5r4KO6YDqwug'
```

```python
customers_df = get_data(customers_url)
orders_df = get_data(orders_url)
order_items_df = get_data(order_items_url)
```

```python
products_merged.head(3)
```

```python
# join orders and items dataframes
orders_items_merged = order_items_df.merge(delivered_orders, how='inner', left_on='order_id', right_on='order_id')
merged_orders_customers = orders_items_merged.merge(customers_df)
merged_orders_customers = merged_orders_customers.merge(products_merged)
```

```python
# add new columns with total price, date, month and year of orders
merged_orders_customers['total_price'] = merged_orders_customers['price'] + merged_orders_customers['freight_value']
merged_orders_customers['purchase_date'] = pd.to_datetime(merged_orders_customers.purchase_timestamp.dt.date)
merged_orders_customers['purchase_month'] = merged_orders_customers.purchase_timestamp.dt.to_period('M')
merged_orders_customers['purchase_year'] = merged_orders_customers.purchase_timestamp.dt.year
```

```python
merged_orders_customers.head(5)
```

# Data Wrangling and Aggregating

```python
# aggregate data by month and calculate revenue, ARPPU, AOV
month_df = merged_orders_customers.groupby('purchase_month')\
    .agg({'customer_unique_id': 'nunique', 'order_id': 'nunique', 'total_price':'sum'})\
    .rename(columns={'customer_unique_id': 'customers_cnt', 'order_id':'orders_cnt', 'total_price': 'revenue'})\
    ['2017':]\
    .assign(ARPPU = lambda x: np.round(x['revenue'] / x['customers_cnt'], 2),
            AOV = lambda x: x['revenue'] / x['orders_cnt']).reset_index()
```

```python
# aggregate data by day and calculate revenue, ARPPU, AOV, add rolling columns to reduce noise
day_df = merged_orders_customers.groupby('purchase_date')\
        .agg({'customer_unique_id': 'nunique', 'order_id': 'nunique', 'total_price':'sum'})\
        .rename(columns={'customer_unique_id': 'customers_cnt', 'order_id':'orders_cnt', 'total_price': 'revenue'})\
        ['2017-01-01':]\
        .assign(
        ARPPU = lambda x: np.round(x['revenue'] / x['customers_cnt'], 2),
        AOV = lambda x: x['revenue'] / x['orders_cnt']
        ).reset_index()
day_df['rolling'] = day_df['revenue'].rolling(7).mean()
day_df['rolling_ARPPU'] = day_df['ARPPU'].rolling(7).mean()
```

```python
# aggregate data by year and calculate revenue, ARPPU, AOV
year_df = merged_orders_customers.loc[merged_orders_customers.purchase_year > 2016].groupby('purchase_year')\
        .agg({'customer_unique_id': 'nunique', 'order_id': 'nunique', 'total_price':'sum'})\
        .rename(columns={'customer_unique_id': 'customers_cnt', 'order_id':'orders_cnt', 'total_price': 'revenue'})\
        .assign(
        LTV = lambda x: np.round(x['revenue'] / x['customers_cnt'], 2),
        AOV = lambda x: np.round(x['revenue'] / x['orders_cnt'], 2)
        ).reset_index()
```

# RFM-segmentation.

```python
# calculate the price for every order
order_price_df = merged_orders_customers.groupby('order_id')['total_price'].sum().to_frame(name='order_price').reset_index()
merged_df = merged_orders_customers.merge(order_price_df,  on='order_id')
```

```python
# select the needed columns
merged_df = merged_df[['customer_unique_id', 'order_id', 'order_price', 'purchase_timestamp']]
merged_df.head()
```

```python
# get monetary dataframe
monetory_df = merged_df.groupby('customer_unique_id', as_index=False)\
    ['order_price'].sum().rename(columns={'order_price': 'monetary'})
monetory_df.head()
```

```python
# get the date out of purchase time
merged_df['purchase_date'] = merged_df['purchase_timestamp'].dt.date
```

```python
# define the minimum and maximum dates
print(f'The difference between minimum and maximum days is {(merged_df['purchase_date'].max()- merged_df['purchase_date'].min()).days}')
```

```python
NOW = merged_df['purchase_date'].max() + timedelta(1)
```

```python
# find the most recent purchase date for customers
merged_df['recency'] = merged_df['purchase_date'].apply(lambda x: (NOW - x).days)
```

```python
merged_df = merged_df.drop(columns='purchase_timestamp')
```

```python
# get recency and frequency for each customer
merged_df = merged_df.groupby('customer_unique_id')\
                    .agg({'recency': lambda x: x.min(),
                    'purchase_date': lambda x: len([d for d in x])})\
                    .reset_index().rename(columns={'purchase_date':'frequency'})
```

```python
# join dataframes to get one RFM
rfm_df = merged_df.merge(monetory_df, on='customer_unique_id')
rfm_df.head()
```

```python
# create a copy of dataframe
rfm_copy = rfm_df.copy()
```

```python
# the distributions of frequency and monetary are very skewed with long tails, the recency is not normal too
fig, axes = plt.subplots(1, 3, figsize=(15,4))
#axes = axes.flatten()
for col, ax in zip(['recency', 'frequency', 'monetary'], axes):
    sns.distplot(rfm_copy[col], ax = ax)
plt.show()
```

```python
# get the segments of recency according to number of days
rfm_copy['recency_seg'] = pd.cut(rfm_copy['recency'], [0 , 90, 180, 1000], labels=[3, 2, 1])
```

```python
# visualize the monetary boxplot by recency segments
rfm_copy.groupby('recency_seg').boxplot(column=['monetary'], layout=(1, 3), figsize=(12, 3));
```

```python
# get the frequency segments
rfm_copy['frequency_seg'] = pd.cut(rfm_copy['frequency'], [0, 1, 4, 100], labels=[1,2,3])
rfm_copy.sort_values('frequency', ascending=False).head(5)
```

```python
# get quantiles of monetary
rfm_copy['monetary'].quantile(np.linspace(0.1, 1, 10))
```

```python
# get the monetary segments
rfm_copy['monetary_seg'] = pd.cut(rfm_copy['monetary'], [0, 100, 400, 200000], labels=[1, 2, 3])
```

```python
# the total segments for each customer
rfm_copy['rfm_total'] = rfm_copy.recency_seg.astype(str)  + rfm_copy.frequency_seg.astype(str) + rfm_copy.monetary_seg.astype(str)
rfm_copy[rfm_copy['frequency'] == 1].shape[0] /rfm_copy.shape[0]
```

```python
# calculate how the revenue depends on each monetary segment 
third = round((rfm_copy.query("monetary_seg == 3").monetary.sum() / rfm_copy.monetary.sum() * 100), 2)
second = round((rfm_copy.query("monetary_seg == 2").monetary.sum() / rfm_copy.monetary.sum() * 100), 2)
first = round((rfm_copy.query("monetary_seg == 1").monetary.sum() / rfm_copy.monetary.sum() * 100), 2)
print(f'The first segment generate {first}% of revenue, the second - {second}% and the third {third}%.')
```

```python
# distinguish different categories for each segment    
seg_map= {
    r'111|112|212|121': 'Lost customers',
    r'232|223|222|122|211|132|221': 'Need Attention',
    r'311|312|321': 'Newcomers',
    r'213|123|313|133|322|123|113': 'Potential champions',
    r'333|233|323|332': 'Champions & Loyal'
}
```

```python
# map every customer to category
rfm_copy['category'] = rfm_copy['recency_seg'].astype(str) + rfm_copy['frequency_seg'].astype(str) + rfm_copy['monetary_seg'].astype(str)
rfm_copy['category'] = rfm_copy['category'].replace(seg_map, regex=True)
rfm_copy.head()
```

```python
# get dataframe with categories and metrics
category_df = rfm_copy.groupby('category', as_index=False).agg({'monetary': 'sum', 'recency':'count'})\
                                .rename(columns={'monetary': 'revenue','recency':'customers_number'})
```

```python
category_df['ARPPU'] = category_df['revenue'] / category_df['customers_number']
```

```python
# checking the number of each category and revenue per user
category_df.sort_values('ARPPU', ascending=False)
```

# Recommendations:
### 1) As our e-commerce business in Acquisition state, we need to aquire new customers and impove our customer service to build customers' trust.
### 2) We need to reduce our delivery time and develop a model to predict delivery time better.
### 3) For loyal and champions category we may give special bonus card with some benefits(price discount, special customer service line).
### 4) For potential champions we need to give them opportunity to buy more frequent or spend more money. We can stimulate their activity by giving them limited in time discounts for some products and categories.
### 5) For Need Attention category we can launch a campaign to reactivate these customers. We can send them e-mail or messanges and give them special coupons or promocodes, we can promise 5-10% cashback to make meth potentially loyal.
### 5) As for Lost customers, it is better not to invest in this category because of lack of returns. However, we can try a small group and reactivate them as needs attention category as an experiment.
